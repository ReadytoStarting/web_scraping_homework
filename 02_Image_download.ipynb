```python
import requests as req
import os
from bs4 import BeautifulSoup

def download_one_episode(title, no, url):
    req_header = {
        'referer': url
    }
    res = req.get(url)
    if res.ok:
        #jpg ÌååÏùºÏùò Ï†àÎåÄÍ≤ΩÎ°ú urlÏùÑ Ï∞æÍ∏∞ 
        soup = BeautifulSoup(res.text, 'html.parser')

        print(len(soup.select("img[src*='IMAG01']")))
        print(len(soup.select("img[src$='.jpg']")))#236
        print(len(soup.select("img[src^='https://image-comic']")))#236

        img_tags = soup.select("img[src*='IMAG01']")
        
        #img srcÎ•º Ï†ÄÏû•Ìï† list ÏÑ†Ïñ∏ 
        img_url_list = []

        for img_tag in img_tags:
            # print(type(img_tag),img_tag.name, img_tag)
            img_src = img_tag['src']
            img_url_list.append(img_src)
        # print(len(img_url_list), img_url_list)

        #img ÎîîÎ†âÌÜ†Î¶¨ ÏóÜÏúºÎ©¥ ÏÉùÏÑ±ÌïòÍ∏∞ 
        imgdir_name = f'img/{title}/{no}'
        if not os.path.isdir(imgdir_name):
            os.makedirs(imgdir_name)

        #os.path.join(dir, filename)

        for img_url in img_url_list:
            i = img_url_list.index(img_url)
        # requests Ïùò get(url, headers) Ìï®Ïàò Ìò∏Ï∂úÌïòÍ∏∞ 
            res = req.get(img_url, headers=req_header)
            # print(res.status_code)
            if res.ok:
                # binary ÏùëÎãµ Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞ content ÌîÑÎ°úÌçºÌã∞Î•º ÏÇ¨Ïö©Ïö©
                img_data = res.content
                # print(img_data)
                # urlÏóêÏÑú ÌååÏùºÎ™ÖÎßå Ï∂îÏ∂úÌïòÍ∏∞
                file_name = os.path.basename(img_url)
                # print(file_name)
                file_name = f"{no}_{i+1}.jpg"
                file_name = os.path.join(imgdir_name, file_name)
                #file_name = os.path.join(imgdir_name, os.path.basename(img_url))
                # binday dataÎ•º fileÏóê writeÌïòÍ∏∞
                with open(file_name, "+wb") as file:
                    print(f'writing to {file_name}({len(img_data):,} bytes)')
                    
                    file.write(img_data)

    else:
        print(f'Error Code = {res.status_code}')
```

```python
download_one_episode('ÏåâÏ¥àÏùò ÎÇú',108,'https://comic.naver.com/webtoon/detail?titleId=823195&no=108&week=wed')
```

```python
import requests
def upload_img(title, no):
    folder_path = os.path.join("img", title, str(no))

    if not os.path.isdir(folder_path):
        print(f"‚ùå Ìè¥ÎçîÍ∞Ä Ï°¥Ïû¨ÌïòÏßÄ ÏïäÏïÑÏöî: {folder_path}")

    upload_files = []
    for file_name in sorted(os.listdir(folder_path)):
        if file_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            file_path = os.path.join(folder_path, file_name)
            upload_files.append(('img1',open(file_path,'rb')))

    
    print(f"üì§ ÏóÖÎ°úÎìúÌï† ÌååÏùº: {[f[1].name for f in upload_files]}")

    url = 'http://httpbin.org/post'
    # file ÏóÖÎ°úÎìú ÌïòÎ†§Î©¥ requestsÏùò post Ìï®ÏàòÏóê files ÏÜçÏÑ±ÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
    res = requests.post(url, files=upload_files)
    print(res.status_code)
    #print(res.json())
    print(res.json()['files'])

    for _,file in upload_files:
        file.close()
```
