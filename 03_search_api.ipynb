```python

# 'data/nhnblog.txt' 파일 생성하기
import re
import os
import json
import requests
from pprint import pprint 
from dotenv import load_dotenv


# .config 파일에서 환경 변수 로드
with open('.config', 'r', encoding='utf-8') as f:
    config = json.load(f)

# 환경 변수에서 값 읽기
headers = {
    "X-Naver-Client-Id" : config["Naver_Developer"]['client_id'],
    "X-Naver-Client-Secret" : config["Naver_Developer"]['client_secret']
}

def search(type, query):
    
    #query string 문자열을 dict 선언
    payload = {
        'query': query,
        'display': 50,
        'sort': 'sim'
    }

    url = f'https://openapi.naver.com/v1/search/{type}.json' #?query=파이썬&display=100&sort=sim


    # requests get(url, params, headers) 요청 
    res = requests.get(url, params=payload, headers=headers)
    # json() 함수로 응답 결과 가져오기
    items_data = res.json()['items']

    # print(type(res.json()))
    # print(len(items_data), type(items_data[0]))
    #pprint(items_data)

    #정규표현식 사용 -> \n\r\t 제거거
    pattern = re.compile(r'[\n\r\t]')

    item_name = list({key for item in items_data for key in item.keys()})
    # [{},{}] 
    item_lists = list()
    item_list = dict()

    # pprint(item_name)
    for item in items_data:
    #     print(item)
            # 'title' , 'description' , 'discount' , 'image' , 'link'
        for key in item_name:
            item_list[key] = pattern.sub("",item[key])

        item_lists.append(item_list)
        item_list = {}


    # pprint(item_lists)#[[],[]]
    global typedir_name
    typedir = f'data/{type}_{query}'
    filname = f'{query}.json'
    if not os.path.isdir(typedir):
        os.makedirs(typedir)
    typedir_name = f'{typedir}/{filname}'
    # 'data/nhnblog.txt' 파일 생성하기
    with open(typedir_name,'w', encoding='utf-8')as json_file:
        json.dump(item_lists, json_file, ensure_ascii=False, indent=4 )

search('book', 'python')
```

```python
import pandas as pd

book_lists = pd.read_json(typedir_name)
book_lists.head()
```

```python
book_lists.loc[book_lists['discount'] >= 20000, book_lists.columns.drop(['description', 'image', 'isbn', 'link'])].sort_values(['discount'],ascending=False).reset_index(drop=True)
```

```python
book_lists.loc[book_lists['publisher'] == "인피니티북스", book_lists.columns.drop(['image','description'])].reset_index()
```
